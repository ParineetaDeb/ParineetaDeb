# **9. Outlier detection **

```python
# === Step 9: Outlier Detection ===
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. Select column for outlier detection ---
col = 'additions'  # change to any numeric column

# --- 2. Calculate IQR ---
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# --- 3. Identify outliers ---
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
print(f"Number of outliers in '{col}': {len(outliers)}")

# --- 4. Plot Boxplot with Outliers Highlighted ---
plt.figure(figsize=(8, 5))
sns.boxplot(x=df[col], color="lightblue")

# Highlight outliers
plt.scatter(outliers[col], [0] * len(outliers), color='red', s=100, marker='o', label='Outliers')

plt.title(f"Outlier Detection in '{col}' Column", fontsize=14)
plt.legend()
plt.tight_layout()
plt.show()
```

# **Discretization (Binning)**

```python
# We'll discretize (bin) a few continuous numeric features
from sklearn.preprocessing import KBinsDiscretizer

# Select up to first 5 numeric columns for demonstration
cols_to_discretize = X.columns[:5]
print("Columns to discretize:", cols_to_discretize.tolist())

# Create discretizer: 4 bins (quartiles), equal frequency ('quantile')
discretizer = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='quantile')

# Apply discretization
X_disc = X.copy()
X_disc[cols_to_discretize] = discretizer.fit_transform(X[cols_to_discretize])

# Visualize before/after for the first column
feature = cols_to_discretize[0]
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(X[feature], bins=20, kde=True)
plt.title(f"Original Distribution — {feature}")

plt.subplot(1, 2, 2)
sns.histplot(X_disc[feature], bins=4, discrete=True)
plt.title(f"After Discretization — {feature}")

plt.tight_layout()
plt.show()

# Replace X with discretized version for further processing
X = X_disc.copy()
```

# **Normalization **

```python
# =====================================
# STEP 8B: Normalization (Three Methods)
# =====================================

from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1️⃣ Min–Max Scaling (scales values to [0, 1]) ---
minmax_scaler = MinMaxScaler()
X_minmax = pd.DataFrame(minmax_scaler.fit_transform(X),
                        columns=X.columns, index=X.index)

# --- 2️⃣ Z-Score Standardization (mean=0, std=1) ---
zscore_scaler = StandardScaler()
X_zscore = pd.DataFrame(zscore_scaler.fit_transform(X),
                        columns=X.columns, index=X.index)

# --- 3️⃣ Decimal Scaling (move decimal point based on max absolute value) ---
X_decimal = X.copy()
for col in X.columns:
    max_abs = X[col].abs().max()
    if max_abs == 0:
        X_decimal[col] = 0
    else:
        j = np.ceil(np.log10(max_abs + 1))
        X_decimal[col] = X[col] / (10 ** j)

# --- Visualization: Compare one feature across methods ---
feature = X.columns[0]  # choose first numeric feature
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.histplot(X_minmax[feature], bins=20, kde=True, color='skyblue')
plt.title(f'Min–Max Scaling: {feature}')

plt.subplot(1, 3, 2)
sns.histplot(X_zscore[feature], bins=20, kde=True, color='orange')
plt.title(f'Z-Score Scaling: {feature}')

plt.subplot(1, 3, 3)
sns.histplot(X_decimal[feature], bins=20, kde=True, color='limegreen')
plt.title(f'Decimal Scaling: {feature}')

plt.tight_layout()
plt.show()

# --- Choose one normalization for modeling (you can switch as desired) ---
# Uncomment ONE line below depending on which normalization you want to use:

# X = X_minmax.copy()   # Min–Max Scaling
# X = X_zscore.copy()   # Z-Score Standardization
X = X_decimal.copy()    # Decimal Scaling (currently active)
```

# **DIMENSION REDUCTION USING LDA**

```python
# =====================================
# STEP 8C: Dimensionality Reduction using LDA
# =====================================

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt
import seaborn as sns

# --- Ensure we have numeric data and target variable ---
if 'y' not in locals():
    raise NameError("Target variable 'y' not found — run target creation step first.")

# --- Initialize LDA ---
lda = LDA(n_components=1)  # You can choose 1, 2, or more components (<= n_classes - 1)

# Fit LDA on full dataset (supervised reduction)
X_lda = lda.fit_transform(X, y)

# --- Convert to DataFrame for visualization ---
lda_df = pd.DataFrame(X_lda, columns=['LDA1'])
lda_df['risky_change'] = y.values

# --- Explained variance ratio ---
print("Explained Variance Ratio by LDA Components:", lda.explained_variance_ratio_)

# --- Visualization of LDA projection ---
plt.figure(figsize=(8, 5))
sns.histplot(
    data=lda_df, x='LDA1', hue='risky_change',
    kde=True, palette='viridis', element='step'
)
plt.title('LDA Projection — Risky vs Safe Code Changes')
plt.xlabel('LDA Component 1')
plt.ylabel('Density')
plt.show()

# --- Optional: If you want 2D visualization (if n_components ≥ 2) ---
# lda = LDA(n_components=2)
# X_lda_2d = lda.fit_transform(X, y)
# lda2_df = pd.DataFrame(X_lda_2d, columns=['LDA1', 'LDA2'])
# lda2_df['risky_change'] = y.values
# plt.figure(figsize=(7,6))
# sns.scatterplot(data=lda2_df, x='LDA1', y='LDA2', hue='risky_change', palette='coolwarm', alpha=0.7)
# plt.title('2D LDA Projection')
# plt.show()

# --- Replace X with LDA-transformed data for modeling ---
X = pd.DataFrame(X_lda, columns=['LDA1'], index=y.index)
```

# **PERFORMANCE ANALYSIS**

```python
# =====================================
# STEP 13: Predictive Modeling & Evaluation
# =====================================

# --- Imports ---
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.metrics import (
    accuracy_score, confusion_matrix, classification_report,
    f1_score, mean_squared_error, silhouette_score, adjusted_rand_score
)
from scipy.stats import mode
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --- Train-test split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# --- 1️⃣ Logistic Regression ---
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_log = log_reg.predict(X_test)

# --- 2️⃣ Random Forest ---
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# --- 3️⃣ Support Vector Machine (SVM) ---
svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)

# --- 4️⃣ Unsupervised: K-Means Clustering ---
# Fit on full dataset, compare clusters to actual labels
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X)
cluster_labels = kmeans.labels_

# Map cluster labels to majority class
mapping = {}
for i in range(2):
    mask = (cluster_labels == i)
    majority = mode(y[mask], keepdims=True).mode[0]  # ✅ Fixed line
    mapping[i] = majority

y_pred_kmeans = np.array([mapping[label] for label in cluster_labels])

# =============================
# 🔹 EVALUATION SECTION
# =============================

# --- Logistic Regression ---
print("\n=== Logistic Regression ===")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print("F1-Score:", f1_score(y_test, y_pred_log))
sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt="d", cmap="Blues")
plt.title("Logistic Regression - Confusion Matrix")
plt.show()

# --- Random Forest ---
print("\n=== Random Forest ===")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("F1-Score:", f1_score(y_test, y_pred_rf))
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt="d", cmap="Greens")
plt.title("Random Forest - Confusion Matrix")
plt.show()

# --- Support Vector Machine ---
print("\n=== Support Vector Machine (SVM) ===")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print("F1-Score:", f1_score(y_test, y_pred_svm))
sns.heatmap(confusion_matrix(y_test, y_pred_svm), annot=True, fmt="d", cmap="Oranges")
plt.title("SVM - Confusion Matrix")
plt.show()

# --- K-Means ---
print("\n=== K-Means Clustering (Unsupervised) ===")
print("Adjusted Rand Index:", adjusted_rand_score(y, y_pred_kmeans))
print("Silhouette Score:", silhouette_score(X, kmeans.labels_))

# =============================
# 🔹 RMSE for Probabilistic Models
# =============================
rmse_log = np.sqrt(mean_squared_error(y_test, log_reg.predict_proba(X_test)[:, 1]))
rmse_rf = np.sqrt(mean_squared_error(y_test, rf.predict_proba(X_test)[:, 1]))
rmse_svm = np.sqrt(mean_squared_error(y_test, svm.predict_proba(X_test)[:, 1]))

print("\n=== RMSE Comparison ===")
print(f"Logistic Regression RMSE: {rmse_log:.4f}")
print(f"Random Forest RMSE: {rmse_rf:.4f}")
print(f"SVM RMSE: {rmse_svm:.4f}")

# =============================
# 🔹 Summary Table
# =============================

results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'SVM', 'K-Means (Unsupervised)'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_log),
        accuracy_score(y_test, y_pred_rf),
        accuracy_score(y_test, y_pred_svm),
        np.nan  # not applicable for unsupervised
    ],
    'F1-Score': [
        f1_score(y_test, y_pred_log),
        f1_score(y_test, y_pred_rf),
        f1_score(y_test, y_pred_svm),
        np.nan
    ],
    'RMSE': [rmse_log, rmse_rf, rmse_svm, np.nan],
    'Adjusted Rand Index': [np.nan, np.nan, np.nan, adjusted_rand_score(y, y_pred_kmeans)],
    'Silhouette Score': [np.nan, np.nan, np.nan, silhouette_score(X, kmeans.labels_)]
})

print("\n=== 📊 Model Performance Summary ===")
display(results)

# --- Bar Plot for Visual Comparison ---
plt.figure(figsize=(8, 5))
sns.barplot(x='Model', y='Accuracy', data=results, palette='viridis')
plt.title("Model Accuracy Comparison")
plt.xticks(rotation=25)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```
